```python
import matplotlib.pyplot as plt
import plotting_utils
import seaborn as sns
import numpy as np

plotting_utils.setup_plotting_params()
import math
 
time_list = [173.2441394329071, 218.0188171863556, 266.99483346939087, 312.46169877052307, 358.94810962677, 402.530179977417, 449.6803216934204, 499.3738114833832, 542.9204573631287, 593.1955976486206, 640.4044270515442, 684.7485768795013, 735.2807495594025, 795.2846944332123, 845.9745187759399, 872.8441681861877, 914.7777645587921, 966.357670545578, 998.3034977912903, 1035.1045076847076, 1083.9593777656555, 1142.4443309307098, 1187.898689031601, 1228.125500679016, 1257.3267064094543, 1299.889087677002, 1350.5871665477753, 1402.4881932735443, 1444.7553634643555, 1505.0754182338715, 1533.1851506233215, 1578.285204410553]

def Log2(x):
    if x == 0:
        return False
    return (math.log10(x) / math.log10(2))

def isPowerOfTwo(n):
    return (math.ceil(Log2(n)) == math.floor(Log2(n)))


x = [1,2,4,8,16,32]
time_list_ = [sum(time_list[:i]) / (23 * i) for i in range(1,len(time_list)+1) if isPowerOfTwo(i)]

plt.figure(figsize=(10,5))
plt.plot(x, time_list_, '-', marker='o', label='MGit')
plt.xticks(x)
plt.xlabel("Number of models")
plt.ylabel("Average runtime\n(seconds)")
plt.xscale('log',base=2)
plt.ylim(0, None)
plt.xticks(x, [str(each_x*23) for each_x in x])
plt.legend(loc="upper left", frameon=False)
sns.despine()

plt.savefig("figures/average_insertion_time.pdf", bbox_inches='tight')
plt.show()
```
