```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Custom legend handler
from matplotlib.legend_handler import HandlerLine2D

class HandlerOnlyMarkers(HandlerLine2D):
    def create_artists(self, legend, orig_handle, xdescent, ydescent, width, height, fontsize, trans):
        # Call the parent method to get a list of artists
        artists = super(HandlerOnlyMarkers, self).create_artists(legend, orig_handle, xdescent, ydescent, width, height, fontsize, trans)
        # Apply 'set_linestyle' to each artist in the list
        for artist in artists:
            artist.set_linestyle('None')
            artist.set_markersize(15)      # Increase marker size
        return artists
```

```python
# Define the specific X-axis ticks
x_ticks = [4, 8, 16, 32, 64, 128]

# Create equally spaced values for plotting
x_values_equally_spaced = np.arange(len(x_ticks))  # This creates an array [0, 1, 2, ..., 6]

y_values1 = [
1.090610757,
1.045673716,
1.018787444,
1.003508784,
0.9796650537,
0.9663817859,]
y_values2 = [
1.051948886,
1.023297954,
0.9950810207,
0.9762133574,
0.9661088227,
]
y_values3 = [
1.02231859,
0.9898573848,
0.9755684599,
0.965724475,
]
y_values4 = [
0.9937423887,
0.9791022394,
0.9663291294,
]

# Increasing the font size for the plot elements
plt.rcParams.update({'font.size': 42})  # Update the global font size
# Creating the plot
plt.figure(figsize=(10, 6))
plt.plot(x_values_equally_spaced, y_values1, color='tab:blue', label='seq_len = 32', linestyle='-', marker='o')
plt.plot(x_values_equally_spaced[:5], y_values2, color='tab:red', label='seq_len = 64', linestyle='-', marker='o')
plt.plot(x_values_equally_spaced[:4], y_values3, color='tab:orange', label='seq_len = 128', linestyle='-', marker='o')
plt.plot(x_values_equally_spaced[:3], y_values4, color='tab:green', label='seq_len = 256', linestyle='-', marker='o')

# Setting the Y-axis range and ticks
# Setting the Y-axis range and ticks
plt.ylim(.8, 1.6)
plt.yticks(np.arange(.8, 1.61, .2))

# Setting the X-axis ticks to the specific values with equal spacing
plt.xticks(x_values_equally_spaced, x_ticks)

# Adding labels and title
plt.xlabel('Batch Size')
plt.ylabel('')
#plt.title('Speed Up (Sequential / Collocated (30%))')

# Adding a grid and a legend
#plt.grid(True, linestyle='--', alpha=0.5)

# Adding a grid and a legend with 2 columns
plt.legend(ncol=2, handler_map={plt.Line2D: HandlerOnlyMarkers()},fontsize=24)

# Display the plot
plt.savefig("./figures/collocation/collocation_30.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
